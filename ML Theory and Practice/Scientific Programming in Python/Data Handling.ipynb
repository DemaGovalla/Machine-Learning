```python
import numpy as np
import matplotlib.pyplot as plt

iris_data = np.loadtxt('irisdata.csv',delimiter=',') # Loading iris dataset
print(iris_data.shape)
```

```python

features = iris_data[:,0:4] # Seperating features and target
target = iris_data[:,4]

# feature_names = np.genfromtxt('feature_names.txt', \
#                               delimiter=',', dtype='str') # Loading feature names of iris dataset
# feature_names = np.delete(feature_names,-1)
# target_names = np.genfromtxt('target_names.txt', \
#                              delimiter=',', dtype='str') # Loading target names of iris dataset
print(target)
print(features)
```

```python
# Sum of all sepal and pedal widths 

print(features[:, [1,3]].sum(axis = 0))
```

```python
# plot the average and std of the feature vectors 

fig, ax = plt.subplots()

ax.bar(np.arange(1,5), features.mean(axis=0))
ax.errorbar(np.arange(1, 5), features.mean(axis = 0), features.std(axis = 0), linestyle = "None", marker = 'o', c = 'r')
ax.legend()
plt.xlabel("features")
plt.ylabel("Mean and Standard diviation")
plt.show()

```

```python
# Plotting the scatter plot of the data
plt.scatter(features[:50,0],features[:50,1], \
            s=10, c='r', label = 'setosa')
plt.scatter(features[50:100,0],features[50:100,1], \
            s=10, c='g', label = 'versicolor')
plt.scatter(features[100:,0],features[100:,1], \
            s=10, c='b', label = 'verginica')
plt.legend(loc='upper right')
plt.xlabel('Sepal length in cm')
plt.ylabel('Sepal width in cm')
plt.show()
```
